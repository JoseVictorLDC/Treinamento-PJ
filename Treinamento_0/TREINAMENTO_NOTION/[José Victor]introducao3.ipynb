```python
# Geral
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Disciplinas_bixos.xlsx', sheet_name='Disciplinas')
df
```

### 1. Quais são as matérias com maior e menor taxa de reprovação?

```python
# Matéria com mais aprovação
df[df["Nº de Aprovações"] == df["Nº de Aprovações"].max()]["Matéria"]
```

```python
# Matéria com menos aprovação
df[df["Nº de Aprovações"] == df["Nº de Aprovações"].min()]["Matéria"]
```

```python
# Analise gráfica
dfReprovacao = df
dfReprovacao["Taxa de reprovação"] = dfReprovacao["Nº de Reprovações"] / dfReprovacao["Nº de Alunos inscritos"]

# Configurando variaveis do gráfico
plt.figure(figsize=(15, 7))
y = dfReprovacao["Taxa de reprovação"]
x = dfReprovacao["Matéria"]

plt.ylabel('TAXA DE REPROVAÇÃO')
plt.xlabel('MATÉRIA')
plt.title('Matérias com maior e menor taxa de reprovação')

plt.xticks(rotation=45, ha='right')

# Plotando gráfico
plt.bar(x,y)
plt.show()
```

### INSIGHT OBJETIVO

As materias com maior e menor taxa de reprovação são respectivamente "Introdução a engenharia mecatrônica" e "Introdução e economia", que pertencem respectivamente a engenharia mecatrônica e engenharia de produção.

É valido destacar que essa analise de taxa de reprovação é oposta a analise que será feita mais a frente da porcentagem de aprovados, no entanto nesse caso será exclusivo para disciplinas específicas.

### 2. Como varia a relação entre o número de professores em relação ao de alunos inscritos por disciplina?

```python
dfProfessores = df
dfProfessores["Relação professores/alunos"] = dfProfessores["Nº de Professores"] / dfProfessores["Nº de Alunos inscritos"]
dfProfessores
```

```python
# Analise gráfica
# Configurando variaveis do gráfico
plt.figure(figsize=(10, 7))
y = dfProfessores["Nº de Professores"]
x = dfProfessores["Nº de Alunos inscritos"]

plt.ylabel('NÚMERO DE PROFESSORES')
plt.xlabel('NÚMERO DE ALUNOS')
plt.title("Relação professores/alunos")

# Plotando gráfico
plt.scatter(x,y)
plt.show()
```

```python
# Analise das disciplinas com menos alunos

# Filtrando menores números de alunos na disciplina
dfProfessores[dfProfessores["Nº de Alunos inscritos"] < 200]["Engenharia"].value_counts()
```

```python
# Analise das disciplinas com mais alunos

# Filtrando maior números de alunos na disciplina
dfProfessores[dfProfessores["Nº de Alunos inscritos"] > 700]["Engenharia"].value_counts()
```

### INSIGHT OBJETIVO

Analisadno o gráfico observa-se uma quantidade predominante de disciplinas com 800 alunos, o que pode revelar uma certa constante de valor máximo para matéria. Além disso, por muitas delas possuirem esse valor entende-se que existe certa lotação nessas matérias, o que pode revelar dificuldade da disciplina ou a própria superlotação da universidade.

Por outro lado, também se observa uma concentração de disciplinas com menos alunos. Vale mencionar que essas turmas possuem poucos professores e isso pode levar à conclusão de uma possivel facilidade da matérias ou uma especificidade dessas matérias à respectivas engenharias (menos alunos totais). Essa última conclusão é comprovada pelas analises feitas após o gráfico: disciplinas com mais alunos são gerais e disciplinas com menos alunos são específicas.

### 3. Qual engenharia tem mais % de aprovação em sua disciplina específica?

```python
# Criando coluna de porcentagem de aprovação
dfAprovacao = df
dfAprovacao["Porcentagem de aprovação"] = (dfAprovacao["Nº de Aprovações"] / dfAprovacao["Nº de Alunos inscritos"]) * 100

# Retirando as linhas de disciplinas gerais
dfEspecificas = dfAprovacao[dfAprovacao["Engenharia"] != "Geral"]

# Verificando qual é a maior porcentagem e qual é a engenharia correspondente
dfEspecificas[dfEspecificas["Porcentagem de aprovação"] == dfEspecificas["Porcentagem de aprovação"].max()]["Engenharia"]
```

```python
# Analise gráfica
# Configurando variaveis do gráfico
plt.figure(figsize=(6, 7))
y = dfAprovacao[dfAprovacao["Engenharia"] != "Geral"]["Porcentagem de aprovação"]
x = dfAprovacao[dfAprovacao["Engenharia"] != "Geral"]["Engenharia"]

plt.ylabel('PORCENTAGEM DE APROVAÇÃO')
plt.xlabel('DISCIPLINA')
plt.title("Porcentagem de aprovação em disciplina específica")

plt.xticks(rotation=45, ha='right')

# Plotando gráfico
plt.bar(x,y)
plt.show()
```

### INSIGHT OBJETIVO

Como que se pode obervar no gráfico acima, entende que a disciplina específica de engenharia de produção é a com maior taxa de aprovação. Isso pode revelar alguns pressupsotos sobre a engenharia ou sua disciplina específica que deveriam ser mais cuidadosamente analisados, dentre eles está uma certa menor cobrança na disciplina ou uma menor carga horária da engenharia de produção, o que pode acarretar em maiores tempos de estudo e maiores notas.

### 4. Qual das disciplinas gerais contribui mais com o número de RECs?

```python
# Retirando as linhas de disciplinas gerais
dfGerais = df[df["Engenharia"] == "Geral"]

# Encontrando a engenharia com maior número de RECs
dfGerais[dfGerais["Nº de RECS"] == dfGerais["Nº de RECS"].max()]["Matéria"]
```

```python
# Analise gráfica
# Configurando variaveis do gráfico
plt.figure(figsize=(10, 10))
x = dfGerais["Nº de RECS"]

plt.title("Disciplinas gerais que mais com o número de RECs")

# Plotando gráfico
plt.pie(x, labels=dfGerais["Matéria"], autopct='%1.1f%%')
plt.show()
```

### INSIGHT OBJETIVO

Partindo do pressuposto de uma conhecida dificuldade das matérias de algebra linear, cálculo diferencial e física nas universidades brasileiras, o gráfico apenas comprova tais matérias como sendo, em destaeue, as que mais levam alunos à recuperação também na escola politécnica.

### 5. Como é a distribuição na Poli de alunos por Engenharia?

```python
# Criando data frame para agrupar engenharias
dfEngenharias = df[df["Engenharia"] != "Geral"] # Retirando as linhas de disciplinas gerais

# Configurando variaveis do gráfico
plt.figure(figsize=(15, 7))
x = dfEngenharias["Engenharia"]
y = dfEngenharias["Nº de Alunos inscritos"]

plt.ylabel('NÚMERO DE ALUNOS')
plt.xlabel('ENGENHARIAS')
plt.title('Distribuição na Poli de alunos por Engenharia')

# Plotando gráfico
plt.bar(x,y)
plt.show()
```

### INSIGHT OBJETIVO

Tendo com base estudos sobre a formação de engenheiros no brasil, entende-se que a maior presença de engenheiros elétricos e engenheiros civis decorrem de uma maior presença dos alunos nesse cursos, o que não é difente na escola Politécnica da USP.
